```python
import cv2
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

```python
# Loading the image
image_path = "image.jpg"  
original_image = cv2.imread(image_path)

# Converting the image from BGR to RGB
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# Displaying the original image
plt.imshow(original_image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

Pencil Sketch Effect

```python
def pencil_sketch(image_rgb):
    # Convert to grayscale
    gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

    # Inverting the grayscale image
    inverted_gray_image = 255 - gray_image

    # Applying Gaussian blur
    blurred_inverted_image = cv2.GaussianBlur(inverted_gray_image, (111, 111), 0)

    # Inverting the blurred image
    inverted_blurred_image = 255 - blurred_inverted_image

    # Creating the pencil sketch effect by mixing the grayscale image with the inverted blurred image
    sketch_image = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)

    return sketch_image

# Applying pencil sketch effect
sketched_image = pencil_sketch(original_image_rgb)

# Displaying the pencil sketch image
plt.imshow(sketched_image, cmap='gray')
plt.title("Pencil Sketch Effect")
plt.axis("off")
plt.show()
```

Sepia Effect

```python
def apply_sepia(image_rgb):
    # Creating the transformation matrix for sepia effect
    kernel = np.array([[0.393, 0.769, 0.189],
                       [0.349, 0.686, 0.168],
                       [0.272, 0.534, 0.131]])

    sepia_image = cv2.transform(image_rgb, kernel)
    sepia_image = np.clip(sepia_image, 0, 255)  
    return sepia_image.astype(np.uint8)
```

```python
# Apply Sepia effect
sepia_image = apply_sepia(original_image_rgb)

# Display the sepia image
plt.imshow(sepia_image)
plt.title("Sepia Effect")
plt.axis("off")
plt.show()
```

Emboss Effect

```python
def emboss_effect(image_rgb):
    # Enhanced kernel for a more prominent emboss effect
    kernel = np.array([[-5, -3, 0], [-3, 2, 3], [0, 3, 5]])
    embossed_image = cv2.filter2D(image_rgb, -1, kernel)
    return embossed_image

# Applying the enhanced emboss effect
embossed_image = emboss_effect(original_image_rgb)

# Displaying the embossed image
plt.imshow(embossed_image)
plt.title("Emboss Effect")
plt.axis("off")
plt.show()
```

Grayscale Effect

```python
def apply_grayscale(image_rgb):
    gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
    return gray_image

grayscale_image = apply_grayscale(original_image_rgb)

# Display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

Negative Effect

```python
def apply_negative(image_rgb):
    negative_image = cv2.bitwise_not(image_rgb)
    return negative_image

# Apply the negative effect
negative_image = apply_negative(original_image_rgb)

# Display the negative image
plt.imshow(negative_image)
plt.title("Negative Effect")
plt.axis("off")
plt.show()
```
